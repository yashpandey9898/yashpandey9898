# Problem statement: Number Recognition


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.datasets import mnist
(x_train,y_train),(x_test,y_test) = mnist.load_data()
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
plt.figure(figsize=(25,25))
for i in range(49):
    plt.subplot(7,7,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i],'gray')
    plt.title(y_train[i],fontsize=20)
plt.show()
```

```python
x_train = x_train[:5000,:,:]
x_test = x_test[:3000,:,:]
y_train = y_train[:5000]
y_test = y_test[:3000]

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Normalize the datasets

```python
x_train = x_train/255 
x_test  = x_test/255
```

### Reshaping

```python
x_train = x_train.reshape(5000,28,28,1)
x_test = x_test.reshape(3000,28,28,1)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Import tensorflow libraries

```python
from tensorflow.keras.layers import Conv2D,Flatten,Dense,MaxPooling2D
from tensorflow.keras.models import Sequential
```

```python
y_train[:10]
```

### CNN Architecture

```python
m1 = Sequential()
m1.add(Conv2D(64,(3,3),activation='relu',input_shape=(28,28,1)))
m1.add(MaxPooling2D(pool_size=(2, 2)))

m1.add(Conv2D(64,(3,3),activation='elu'))
m1.add(MaxPooling2D(pool_size=(2, 2)))
       
m1.add(Conv2D(32,(3,3),activation='elu'))
m1.add(MaxPooling2D(pool_size=(2, 2)))    

m1.add(Flatten())
m1.add(Dense(32,activation='elu'))
m1.add(Dense(10,activation='softmax'))

m1.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['acc'])
```

```python
h1 = m1.fit(x_train,y_train,epochs=15,validation_data=(x_test,y_test))
```

```python
r1 = pd.DataFrame(h1.history)
r1['Epochs'] = h1.epoch
r1.tail()
```

```python
plt.plot(r1['Epochs'],r1['loss'],label='Train_loss')
plt.plot(r1['Epochs'],r1['val_loss'],label='Test_loss')
plt.title('Loss vs val_loss')
plt.xlabel('Loss')
plt.ylabel('val_loss')
plt.legend()
plt.show()
```

```python
plt.plot(r1['Epochs'],r1['acc'],label='Train_Acc')
plt.plot(r1['Epochs'],r1['val_acc'],label='Test_Acc')
plt.title('Accuracy vs val_accuracy')
plt.xlabel('Accuracy')
plt.ylabel('val_accuracy')
plt.legend()
plt.show()
```

```python
ypred = m1.predict(x_test)
print(ypred[:10])
```

```python
ypred_m1 = [np.argmax(i) for i in ypred]
print(ypred_m1[:49])
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,ypred_m1))
print(classification_report(y_test,ypred_m1))
```
